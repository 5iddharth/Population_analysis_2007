```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('countries-checkpoint.csv')
df.head(3)
```

```python
year_52 = df.loc[df['year'] == 1952]
year_52.head()
```

```python
year_07 = df.loc[df['year'] == 2007]
year_07.head()
```

```python
type(df), type(year_52), type(year_07)
```

```python
df_01 = year_52.merge(year_07, left_on='country', right_on='country')
df_01
```

```python
df_01.drop(['year_x', 'year_y'], axis = 1, inplace=True)
```

```python
col = ['Country', 'Year-1952', 'Year-2007']
df_01.columns = col
df_01.head(6)
```

```python
df_01.sort_values('Year-2007', ascending = False, inplace=True)
```

```python
country_list = ['China', 'India', 'United States', 'Indonesia', 'Brazil']

pop_growth = (df_01['Year-2007']-df_01['Year-1952']) 

df_01['Growth'] = pop_growth 
df_01

df_01.reset_index(inplace = True)
df_01.drop('index', axis=1, inplace = True)
df_01
```

```python
type(pop_growth), type(df_01['Growth'])
```

## Top 5 Most populated Countries

```python
grow = df_01['Growth']
type(grow)
```

```python
df_01.head(10)
```

```python
df_02 = df_01.loc[:5,:]
df_02
```

```python
plt.figure(figsize= (15,10))
plt.bar(df_02['Country'], df_02['Growth'], color = 'orange')

plt.xlabel('Country')
plt.ylabel('Population Growth (Millions)')
plt.title('Top 10 Countries w/the Biggest Population Growth from 1952 to 2007')

for x,y in zip(df_02['Country'], df_02['Growth']):

    label = "{:.2f}".format(y)

    plt.annotate(label, # this is the text
                 (x,y), # this is the point to label
                 textcoords="offset points", # how to position the text
                 xytext=(0,10), # distance from text to points (x,y)
                 ha='center') # horizontal alignment can be left, right or center

plt.show()
```
